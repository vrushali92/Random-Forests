```python
import pandas as pd
import numpy as np

# resampling techniques
from sklearn.utils import resample

# train-test split
from sklearn.model_selection import train_test_split

# MinMaxscaler
from sklearn.preprocessing import MinMaxScaler

# OneHotEncoder
from sklearn.preprocessing import OneHotEncoder

pd.set_option('display.max_columns',None)
```

```python
categorical_df = pd.read_csv("categorical.csv")
numerical_df = pd.read_csv("numerical.csv")
target_df = pd.read_csv("target.csv")
```

```python
categorical_df
```

```python
numerical_df
```

```python
target_df
```

```python
display(categorical_df.shape)
display(numerical_df.shape)
display(target_df.shape)
```

# Creating features and target

```python
#features
X = pd.concat([categorical_df, numerical_df], axis = 1)

# Target Variable: Binary Indicator for Response to 97NK Mailing
y = target_df.TARGET_B
```

```python
X
```

```python
y
```

# train - test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 1)
```

```python
display(X_train.shape)
display(y_train.shape)

display(X_test.shape)
display(y_test.shape)
```

# Splitting train-test dataset into numericals and categoricals for scaling and encoding

```python
X_train_num = X_train.select_dtypes(np.number)
X_train_cat = X_train.select_dtypes(object)

X_test_num = X_test.select_dtypes(np.number)
X_test_cat = X_test.select_dtypes(object)
```

```python
display(X_train_num.shape)
display(X_train_cat.shape)

display(X_test_num.shape)
display(X_test_cat.shape)
```

```python
y_train.value_counts()
```

# Scaling train-test dataset using MinMaxscaler()

```python
# fitting the dataset
transform = MinMaxScaler().fit(X_train_num)

# scaling train-test dataset using same scaler

X_train_scaled = pd.DataFrame(transform.transform(X_train_num), columns = X_train_num.columns)
X_test_scaled = pd.DataFrame(transform.transform(X_test_num), columns = X_test_num.columns)
```

```python
display(X_train_scaled.shape)
display(X_test_scaled.shape)
```

# Encoding train-test dataset using OneHotEncoder

```python
# fitting the model
encoder = OneHotEncoder(drop = "first").fit(X_train_cat)

# encoding train-test dataset using same encoder

column_names = encoder.get_feature_names_out(X_train_cat.columns)

X_train_encoded = pd.DataFrame(encoder.transform(X_train_cat).toarray(), columns = column_names)
X_test_encoded = pd.DataFrame(encoder.transform(X_test_cat).toarray(), columns = column_names)
```

```python
display(X_train_encoded.shape)
display(X_test_encoded.shape)
```

# Concatenating the processed train-test dataset

```python
X_train_processed = pd.concat([X_train_encoded, X_train_scaled], axis = 1)
X_test_processed = pd.concat([X_test_encoded, X_test_scaled], axis = 1)
```

```python
display(X_train_processed.shape)
display(X_test_processed.shape)
```

# Concatenating only train data for oversampling technique

```python
# Oversampling/Undersampling only on training data

X_train_processed = X_train_processed.reset_index()
y_train = y_train.reset_index()

alldata = pd.concat([X_train_processed, y_train], axis = 1)
alldata = alldata.reset_index()
alldata
```

```python
alldata = alldata.drop(["level_0", "index"], axis = 1)
alldata
```

```python
alldata.isna().sum()
```

# Oversampling

```python
# As there is large difference between people donated or not[TARGET_B]. Hence need to oversample '1'.
```

```python
alldata.TARGET_B.value_counts()
```

```python
category_0 = alldata[alldata['TARGET_B'] == 0]
category_1 = alldata[alldata['TARGET_B'] == 1]

display(category_0.shape)
display(category_1.shape)
```

```python
# Oversampling category_1
category_1_oversampled = resample(category_1,
                                  replace=True,
                                  n_samples = len(category_0),
                                  random_state = 0)
```

```python
display(category_1_oversampled.shape)
display(category_0.shape)
```

```python
oversampled_data = pd.concat([category_0, category_1_oversampled], axis = 0)
oversampled_data
```

# Now applying RandomForestAlgorithm on oversampled data

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
#randomize the rows
oversampled_data = oversampled_data.sample(frac = 1) 

X_train_over = oversampled_data.drop(["TARGET_B"], axis = 1)
y_train_over = oversampled_data["TARGET_B"]
```

```python
X_train_over
```

```python
y_train_over
```

```python
# RandomForestClassifier is collection of multiple DecisionTreeClassifier.

clf = RandomForestClassifier(max_depth=5, # max number of questions to ask
                             min_samples_split=20, # amount of rows still considered at every question
                             min_samples_leaf =20, # ultimate answer based on at least this many rows
                             max_samples=0.8, # fraction of X-train to use in each tree
                             random_state=42)
clf.fit(X_train_over, y_train_over)

print(clf.score(X_train_over, y_train_over))
print(clf.score(X_test_processed, y_test))
```

```python
y_pred = clf.predict(X_test_processed)

display(y_test.value_counts())
display(confusion_matrix(y_test, y_pred))
```

# Feature Selection - RFE

```python
# concatenating previously encoded and scaled X-train and test.
X = X_train_over
X
```

```python
y = y_train_over
y
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.feature_selection import RFE
from sklearn import linear_model
lm = linear_model.LogisticRegression()
```

```python
rfe = RFE(lm, n_features_to_select = 20, verbose=False)
```

```python
rfe.fit(X, y)
```

```python
print(rfe.ranking_)
len(rfe.ranking_)
```

```python
# After we run the algorithm, it labels the top features as 1 and the rest are marked in an increasing order of importance.
df = pd.DataFrame(data = rfe.ranking_, columns = ['Rank'])
df['Column_name'] = pd.DataFrame(X).columns

selected_features = df[(df['Rank'] == 1)]
display(len(selected_features))

display(selected_features)
```

```python
selectedFeatures = selected_features["Column_name"]
X_rfe = X[selectedFeatures]
```

```python
X_rfe
```

```python
X_rfe.dtypes
```

# After selecting features, now doing train-test split

```python
X_train_rfe, X_test_rfe, y_train_rfe, y_test_rfe = train_test_split(X_rfe, y, test_size=0.25, random_state = 1)
```

# Now applying RandomTreeClassifier on selected features

```python
clf_rfe = RandomForestClassifier(max_depth = 15, 
                                min_samples_split = 20, 
                                min_samples_leaf = 20, 
                                max_samples = 0.8)
```

```python
clf_rfe.fit(X_train_rfe, y_train_rfe)
```

```python
print(clf_rfe.score(X_train_rfe, y_train_rfe))
print(clf_rfe.score(X_test_rfe, y_test_rfe))
```

```python
y_pred_rfe = clf_rfe.predict(X_test_rfe)

display(y_test_rfe.value_counts())
display(confusion_matrix(y_test_rfe, y_pred_rfe))
```

# Variance Threshold

```python
from sklearn.feature_selection import VarianceThreshold 

# We are using variance to remove some of the columns.

var_threshold = 0.02
sel = VarianceThreshold(threshold=(var_threshold))
```

```python
X_var = X_train_over
y_var = y_train_over
```

```python
X_var
```

```python
y_var
```

```python
# X_train_var, X_test_var, y_train_var, y_test_var = train_test_split(X_var, y_var, test_size=0.25)
```

```python
var_threshold = 0.02
sel = VarianceThreshold(threshold=(var_threshold))
```

```python
sel = sel.fit(X_var)
```

```python
sel.get_support()
var_list=list(sel.get_support())
droplist_var=[col[0] for col in zip(X_var.columns, var_list) if col[1] == False]
print(droplist_var)
len(droplist_var)
```

```python
X_var = X_var.drop(droplist_var, axis=1)
X_var
```

```python
X_train_var, X_test_var, y_train_var, y_test_var = train_test_split(X_var, y_var, test_size=0.25)
```

```python
clf_var = RandomForestClassifier(max_depth=10, 
                             min_samples_split=20, 
                             min_samples_leaf =20, 
                             max_samples=0.8)
```

```python
clf_var.fit(X_train_var, y_train_var)
```

```python
print(clf_var.score(X_train_var, y_train_var))
print(clf_var.score(X_test_var, y_test_var))
```

```python
y_pred_var = clf_var.predict(X_test_var)

display(y_test_var.value_counts())
display(confusion_matrix(y_test_var, y_pred_var))
```
